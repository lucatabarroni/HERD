```python
class DataBatcher(Sequence):
    def __init__(self,
                 list_low_elcectrons,list_high_electrons,list_low_protons,list_high_protons,
                 list_low_electrons_size,list_high_electrons_size,list_low_protons_size,list_high_protons_size,
                 massimo,batch_size=64):
        
        self.low_electrons=list_low_elcectrons
        self.high_electrons=list_high_electrons
        self.low_protons=list_low_protons
        self.high_protons=list_high_protons

        self.low_electrons_size=list_low_electrons_size
        self.high_electrons_size=list_high_electrons_size
        self.low_protons_size=list_low_protons_size
        self.high_protons_size=list_high_protons_size

        self.remaining_high_electrons=sum(list_high_electrons_size)
        self.remaining_low_electrons=sum(list_low_electrons_size)
        self.remaining_high_protons=sum(list_high_protons_size)
        self.remaining_low_protons=sum(list_low_protons_size)

        self.massimo=massimo
        self.batch_size=batch_size

        self.dataset=[]

        self.perc_low_electrons=sum(self.low_electrons_size)/(sum(self.low_electrons_size)+sum(self.high_electrons_size)+sum(self.low_protons_size)+sum(self.high_protons_size))
        self.perc_high_electrons=sum(self.high_electrons_size)/(sum(self.low_electrons_size)+sum(self.high_electrons_size)+sum(self.low_protons_size)+sum(self.high_protons_size))
        self.perc_high_protons=sum(self.high_protons_size)/(sum(self.low_electrons_size)+sum(self.high_electrons_size)+sum(self.low_protons_size)+sum(self.high_protons_size))
        self.perc_low_protons=sum(self.low_protons_size)/(sum(self.low_electrons_size)+sum(self.high_electrons_size)+sum(self.low_protons_size)+sum(self.high_protons_size))

        #questa cosa va corretta    
        self.low_electrons_per_batch=self.batch_size*self.perc_low_electrons
        self.high_electrons_per_batch=self.batch_size*self.perc_high_electrons
        self.low_protons_per_batch=self.batch_size*self.perc_low_protons
        self.high_protons_per_batch=self.batch_size*self.perc_high_protons

        self.low_electrons_batch_per_file=[[None, None] for _ in range(len(self.low_electrons))]
        self.high_electrons_batch_per_file=[[None, None] for _ in range(len(self.high_electrons))]
        self.low_protons_batch_per_file=[[None, None] for _ in range(len(self.low_protons))]
        self.high_protons_batch_per_file=[[None, None] for _ in range(len(self.high_protons))]

        initial_low_electrons_batch=0
        initial_high_electrons_batch=0
        initial_hig_protons_batch=0
        initial_low_protons_batch=0

        self.high_ele_to_skip=0
        self.low_ele_to_skip=0
        self.high_pro_to_skip=0
        self.low_pro_to_skip=0

        self.high_ele_to_stop=0
        self.low_ele_to_stop=0
        self.high_pro_to_stop=0
        self.low_pro_to_stop=0
        
        for i in range(len(self.low_electrons)):
            self.low_electrons_batch_per_file[i][0]=initial_low_electrons_batch
            final_low_electrons_batch=initial_low_electrons_batch+(self.low_electrons_size[i]/self.low_electrons_per_batch)
            self.low_electrons_batch_per_file[i][1]=final_low_electrons_batch
            initial_low_electrons_batch=final_low_electrons_batch
        
        for i in range(len(self.high_electrons)):
            self.high_electrons_batch_per_file[i][0]=initial_high_electrons_batch
            final_high_electrons_batch=initial_high_electrons_batch+(self.high_electrons_size[i]/self.high_electrons_per_batch)
            self.high_electrons_batch_per_file[i][1]=final_high_electrons_batch
            initial_high_electrons_batch=final_high_electrons_batch
        
        for i in range(len(self.low_protons)):
            self.low_protons_batch_per_file[i][0]=initial_low_protons_batch
            final_low_protons_batch=initial_low_protons_batch+(self.low_protons_size[i]/self.low_protons_per_batch)
            self.low_protons_batch_per_file[i][1]=final_low_protons_batch
            initial_low_protons_batch=final_low_protons_batch
        
        for i in range(len(self.high_protons)):
            self.high_protons_batch_per_file[i][0]=initial_hig_protons_batch
            final_high_protons_batch=initial_hig_protons_batch+(self.high_protons_size[i]/self.high_protons_per_batch)
            self.high_protons_batch_per_file[i][1]=final_high_protons_batch
            initial_hig_protons_batch=final_high_protons_batch
            
    #devi correggere len    
    def __len__(self):
        return int(self.low_electrons_batch_per_file[-1][1])
    
    def creation_dataset(self,labels):
        matr_batch=[]
        for i in range(int(self.dataset.shape[0]/400)):
            matrix = np.array(self.dataset.iloc[i*400:(i+1)*400]).reshape(20, 20)
            matr_batch.append(matrix)

        self.dataset=[]
        
        input=list(zip(matr_batch,labels))
        random.shuffle(input)
        
        return input
    
    def __getitem__(self,idx):
        
        for low_ele_file in range(len(self.low_electrons)):
            if self.low_electrons_batch_per_file[low_ele_file][0]<=idx<self.low_electrons_batch_per_file[low_ele_file][1]:
                low_ele_file_to_open=low_ele_file
                break
        
        for high_ele_file in range(len(self.high_electrons)):
            if self.high_electrons_batch_per_file[high_ele_file][0]<=idx<self.high_electrons_batch_per_file[high_ele_file][1]:
                high_ele_file_to_open=high_ele_file
                break
        
        for low_pro_file in range(len(self.low_protons)):
            if self.low_protons_batch_per_file[low_pro_file][0]<=idx<self.low_protons_batch_per_file[low_pro_file][1]:
                low_pro_file_to_open=low_pro_file
                break

        for high_pro_file in range(len(self.high_protons)):
            if self.high_protons_batch_per_file[high_pro_file][0]<=idx<self.high_protons_batch_per_file[high_pro_file][1]:
                high_pro_file_to_open=high_pro_file
                break
        
        self.high_ele_to_skip=int((idx-self.high_electrons_batch_per_file[high_ele_file_to_open][0])*self.high_electrons_per_batch)
        self.low_pro_to_skip=int((idx-self.low_protons_batch_per_file[low_pro_file_to_open][0])*self.low_protons_per_batch)
        self.high_pro_to_skip=int((idx-self.high_protons_batch_per_file[high_pro_file_to_open][0])*self.high_protons_per_batch)
        self.low_ele_to_skip=idx*self.batch_size-self.high_ele_to_skip-self.low_pro_to_skip-self.high_pro_to_skip

        #deve prendere tanti eventi quanti sarnno quelli skippati al giro successivo

        self.high_ele_to_stop=int(((idx+1)-self.high_electrons_batch_per_file[high_ele_file_to_open][0])*self.high_electrons_per_batch)
        self.low_pro_to_stop=int(((idx+1)-self.low_protons_batch_per_file[low_pro_file_to_open][0])*self.low_protons_per_batch)
        self.high_pro_to_stop=int(((idx+1)-self.high_protons_batch_per_file[high_pro_file_to_open][0])*self.high_protons_per_batch)
        self.low_ele_to_stop=(idx+1)*self.batch_size-self.high_ele_to_stop-self.low_pro_to_stop-self.high_pro_to_stop
        
        
        if self.low_electrons_size[low_ele_file_to_open]-self.low_ele_to_skip>self.low_electrons_per_batch:
            with up.open(self.low_electrons[low_ele_file_to_open][:-14]) as file:
                low_ele_df = file['showersTree'].arrays('deps2D', library='pd')[self.low_ele_to_skip*400:self.low_ele_to_stop*400]
        if self.high_electrons_size[high_ele_file_to_open]-self.high_ele_to_skip>=self.high_electrons_per_batch:
            with up.open(self.high_electrons[high_ele_file_to_open][:-14]) as file:
                high_ele_df = file['showersTree'].arrays('deps2D', library='pd')[self.high_ele_to_skip*400:self.high_ele_to_stop*400]
        if self.low_protons_size[low_pro_file_to_open]-self.low_pro_to_skip>=self.low_protons_per_batch:
            with up.open(self.low_protons[low_pro_file_to_open][:-14]) as file:
                low_pro_df = file['showersTree'].arrays('deps2D', library='pd')[self.low_pro_to_skip*400:self.low_pro_to_stop*400]
        if self.high_protons_size[high_pro_file_to_open]-self.high_pro_to_skip>=self.high_protons_per_batch:
            with up.open(self.high_protons[high_pro_file_to_open][:-14]) as file:
                high_pro_df = file['showersTree'].arrays('deps2D', library='pd')[self.high_pro_to_skip*400:self.high_pro_to_stop*400]
        
        if self.low_electrons_size[low_ele_file_to_open]-self.low_ele_to_skip<self.low_electrons_per_batch:
            with up.open(self.low_electrons[low_ele_file_to_open][:-14]) as file:
                with up.open(self.low_electrons[low_ele_file_to_open+1][:-14]) as file2:
                    low_ele_df = file['showersTree'].arrays('deps2D', library='pd')[self.low_ele_to_skip*400:]
                    low_ele_df2 = file2['showersTree'].arrays('deps2D', library='pd')[:(self.low_ele_to_stop*400-self.low_ele_to_skip*400)-low_ele_df.shape[0]]
                    low_ele_df=low_ele_df.append(low_ele_df2)
        
        if self.high_electrons_size[high_ele_file_to_open]-self.high_ele_to_skip<self.high_electrons_per_batch:
            with up.open(self.high_electrons[high_ele_file_to_open][:-14]) as file:
                with up.open(self.high_electrons[high_ele_file_to_open+1][:-14]) as file2:
                    high_ele_df = file['showersTree'].arrays('deps2D', library='pd')[self.high_ele_to_skip*400:]
                    high_ele_df2 = file2['showersTree'].arrays('deps2D', library='pd')[:(self.high_ele_to_stop*400-self.high_ele_to_skip*400)-high_ele_df.shape[0]]
                    high_ele_df=high_ele_df.append(high_ele_df2)    

        if self.low_protons_size[low_pro_file_to_open]-self.low_pro_to_skip<self.low_protons_per_batch:
            with up.open(self.low_protons[low_pro_file_to_open][:-14]) as file:
                with up.open(self.low_protons[low_pro_file_to_open+1][:-14]) as file2:
                    low_pro_df = file['showersTree'].arrays('deps2D', library='pd')[self.low_pro_to_skip*400:]
                    low_pro_df2 = file2['showersTree'].arrays('deps2D', library='pd')[:(self.low_pro_to_stop*400-self.low_pro_to_skip*400)-low_pro_df.shape[0]]
                    low_pro_df=low_pro_df.append(low_pro_df2)
        
        if self.high_protons_size[high_pro_file_to_open]-self.high_pro_to_skip<self.high_protons_per_batch:
            with up.open(self.high_protons[high_pro_file_to_open][:-14]) as file:
                with up.open(self.high_protons[high_pro_file_to_open+1][:-14]) as file2:
                    high_pro_df = file['showersTree'].arrays('deps2D', library='pd')[self.high_pro_to_skip*400:]
                    high_pro_df2 = file2['showersTree'].arrays('deps2D', library='pd')[:(self.high_pro_to_stop*400-self.high_pro_to_skip*400)-high_pro_df.shape[0]]
                    high_pro_df=high_pro_df.append(high_pro_df2)

        labels_low_electrons=[1]*(self.low_ele_to_stop-self.low_ele_to_skip)
        labels_high_electrons=[1]*(self.high_ele_to_stop-self.high_ele_to_skip)
        labels_low_protons=[0]*(self.low_pro_to_stop-self.low_pro_to_skip)
        labels_high_protons=[0]*(self.high_pro_to_stop-self.high_pro_to_skip)

        labels=list(np.array(labels_low_electrons+labels_high_electrons+labels_low_protons+labels_high_protons))
        self.dataset=pd.concat([low_ele_df,high_ele_df,low_pro_df,high_pro_df])
        self.dataset.reset_index(drop=True, inplace=True)
        
        return self.creation_dataset(labels)
```
